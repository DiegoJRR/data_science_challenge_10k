```python
import pandas as pd
from simpledbf import Dbf5


def load_db(path: str) -> pd.DataFrame:
    dbf_table = Dbf5(path)
    df = dbf_table.to_dataframe()
    return df


def rename_cols(df: pd.DataFrame, descriptors_path: str) -> pd.DataFrame:
    dbf_table_descriptors = Dbf5(descriptors_path, codec="latin-1")
    descriptors = dbf_table_descriptors.to_dataframe()
    original_cols = df.columns

    cols_changes = {}
    for col in original_cols:
        cols_changes[col] = descriptors[descriptors['CAMPO'] == col]['DESCRIP'].iloc[0]

    return df.rename(columns=cols_changes)


df = load_db("data/cpv2020_municipal_vivienda.dbf")
df = rename_cols(df, "data/Desc_prueba10k.dbf")
df
```
